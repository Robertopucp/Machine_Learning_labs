### Bootstraping

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt
import seaborn as sns
import pyreadr
```

```python
## loading the data
data= pd.read_csv("../data/penn_jae.dat" , sep='\s', engine='python')
data= data[ (data['tg'] == 4) | (data['tg'] == 0) ]
```

```python
# Creating the new variable
data["log(inuidur1)"] = np.log(data["inuidur1"])
```

```python
data
```

```python
# Creating a new variable, group treatment 4
data['T4'] = (data[['tg']]==4).astype(int)
```

```python
data
```

```python
data.drop(columns='muld', inplace=True)
```

```python
data=data.reset_index(drop=True)
```

```python
data
```

```python
# equivalent to sample function in lab
def get_indices(data,num_samples):
    return  np.random.choice(data.index, num_samples, replace=True)
```

```python
# Creating a function that can get estimations from the equation

def get_estimates(data,index):
    
    #Working with the equation log(inuidur1)~T4+ (female+black+othrace+factor(dep)+q2+q3+q4+q5+q6+agelt35+agegt54+durable+lusd+husd)
    
    X = data[['T4','female','black','othrace','dep','q2','q3','q4','q5','q6','agelt35','agegt54','durable','lusd','husd']].loc[index]#Independents variable
    y = data['log(inuidur1)'].loc[index] #Dependent variable
    
    #Making the regressions
    lr = LinearRegression()
    lr.fit(X,y) #Fitting the regression
    intercept = lr.intercept_
    coef = lr.coef_
    return [intercept,coef]

get_estimates(data,np.arange(0,5099))

#Creating a function that can get the coefficients and intercept through bootstraping method
def boot(data,func,R):
    intercept = []#creating an empty list
    coeff_1 = []#creating an empty list
    coeff_2 = []#creating an empty list
    coeff_3 = []#creating an empty list
    for i in range(R):
        intercept.append(func(data,get_indices(data,100))[0])#Fill in the lists
        coeff_1.append(func(data,get_indices(data,100))[1][0]) #Fill in the lists
        coeff_2.append(func(data,get_indices(data,100))[1][1])#Fill in the lists
        coeff_3.append(func(data,get_indices(data,100))[1][2])    #Fill in the lists
    
    #Getting the coefficients and intercept
    intercept_statistics = {'estimated_value':np.mean(intercept),'std_error':np.std(intercept)}   
    coeff_1_statistices = {'estimated_value':np.mean(coeff_1),'std_error':np.std(coeff_1)}   
    coeff_2_statistices = {'estimated_value':np.mean(coeff_2),'std_error':np.std(coeff_2)}
    coeff_3_statistices = {'estimated_value':np.mean(coeff_3),'std_error':np.std(coeff_3)}  
    return {'intercept':intercept_statistics,'coeff_1_statistices':coeff_1_statistices,'coeff_2_statistics':coeff_2_statistices,'coeff_3_statistics':coeff_3_statistices}

results = boot(data,get_estimates,1000)

#Let's print the results
print('Result for intercept ',results['intercept'])
print('Result for coefficient term T4',results['coeff_1_statistices'])
print('Result for coefficient term female',results['coeff_2_statistics'])
print('Result for coefficient term black',results['coeff_3_statistics'])


#lets see what the model predicts
import statsmodels.api as sm

X = data[['T4','female','black','othrace','dep','q2','q3','q4','q5','q6','agelt35','agegt54','durable','lusd','husd']]
y = data['log(inuidur1)']

X = sm.add_constant(X)
results = sm.OLS(y,X).fit()
print(results.summary())
```
