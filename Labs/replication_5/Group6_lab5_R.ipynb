## Bootstrap

The bootstrap can be used to estimate the standard errors of the coefficients from a linear regression fit, or a
confidence interval for that coefficient. The power of the bootstrap lies in the fact that it can be easily applied to a wide range of statistical learning methods. 

In this replication, we use the Pennsylvania re-employment bonus experiment in order to compare treatment group 4 and the control group, for this purpose, we compute the standard errors of 1,000 bootstrap estimates for the $T4$, $female$ and $black$ variables.

## Example 1

First, we resample the data with replacement, and the size of the resample must be equal to the size of the original data set. Then the statistic of interest is computed from the resample from the first step. We repeat this routine many times to get a more precise estimate of the Bootstrap distribution of the statistic.

We create $boot.fn()$ which takes in the $Penn$ data set as well as a set of indices for the observations, and returns the intercept and slope estimates for the linear regression model. We then apply this function to the full set of  5099  observations in order to compute the estimates 16 variables.

Then, with $boot()$ function to compute the standard errors of 1,000 bootstrap estimates for the intercept and slope terms. Note that we are using $log(inuidur1)~T4+ (female+black+othrace+factor(dep)+q2+q3+q4+q5+q6+agelt35+agegt54+durable+lusd+husd$ formula for estimation, so bootstraping is goíng to estimate 16 variables (intercept included). 

```r
boot.fn <- function(Penn, index)
  coef(lm(log(inuidur1)~T4+ (female+black+othrace+factor(dep)+q2+q3+q4+q5+q6+agelt35+agegt54+durable+lusd+husd), data = Penn, subset = index))
```

```r
lm(log(inuidur1)~T4+ (female+black+othrace+factor(dep)+q2+q3+q4+q5+q6+agelt35+agegt54+durable+lusd+husd), data = Penn)
```

```r
boot.fn(Penn, 1:5099)
```

 The `boot.fn()` function can also be used in order to create
 bootstrap estimates for the intercept and slope terms by randomly sampling from among the observations with replacement. Here we give two examples.

```r
sample(1:5099, 5099, replace = T)
```

```r
set.seed(1)
boot.fn(Penn, sample(5099, 5099, replace = T))
boot.fn(Penn, sample(5099, 5099, replace = T))
```

Next, we use the `boot()` function to compute the standard errors of 1,000 bootstrap estimates for the intercept and slope terms.

```r
boot1 = boot(Penn, boot.fn, 1000)
boot1
```

```r
s =summary(
    lm(log(inuidur1)~T4+ (female+black+othrace+factor(dep)+q2+q3+q4+q5+q6+agelt35+agegt54+durable+lusd+husd)), data = Penn
    )$coef
s
```

```r
library(xtable)
table<- matrix(0, 4, 4)
table[1,1]<-  s[1,1]
table[1,2]<-  s[1,2]
table[1,3]<-  boot1[[1]][1]
table[1,4]<-  "0.14902630"

table[2,1]<-  s[2,1]
table[2,2]<-  s[2,2]
table[2,3]<-  boot1[[1]][2]
table[2,4]<-  "0.03607677"

table[3,1]<-  s[3,1]
table[3,2]<-  s[3,2]
table[3,3]<-  boot1[[1]][3]
table[3,4]<-  "0.03462895"

table[4,1]<-  s[4,1]
table[4,2]<-  s[4,2]
table[4,3]<-  boot1[[1]][4]
table[4,4]<-  "0.06072504"
colnames(table)<- c("estimate","standard error","estimate -boot", "standard error-boot")
rownames(table)<- c("intercept", "T4TRUE", "female", "black")
table
```

- $T4$: In the fourth treated group, says that is $7.2\%$ less likely to be unemployed for one more week compared with groups who dont belong to this fourth group. Regarding its standar error, according to ols method, the standar error is $0.0354632588775146$, meanwhile using bootstraping method the standar error is $0.03607677$.

- $female$: Aditionally, women are $12.6\%$ more likely to be unemployed for one more week compared with men. Regarding its standar error, according to ols method, the standar error is $0.0348249327707101$, meanwhile using bootstraping method the standar error is $0.03462895$.

- $black$: Finally, afro american people is $29.37\%$ less likely to be unemployed for one more week compared with hispanic or white people. Regarding its standar error, according to ols method, the standar error is $0.0529755621788769$, meanwhile using bootstraping method the standar error is $0.06072504$.

In this exercise, we included many control variables, but we only are interested at the standar errors of three of them (T4, female and black). In this summary table, we can notice that the estimaded coefficients are the same for the ols regression and bootstrap method, however, but the standard errors estimated with bootstraping are slightly larger than those estimated by ols, except for female variable.

## Causal Tree

#### Explain why we need to partitionate the data in three sets.

The division into three sub-sets of data has the purpose of obtaining more efficient results and falling into overfiting. We separate the data into three: split, estimation and test. First, in the split we find the ideal tree, second, in the ideal tree I look for the heterogeneous effects of the treatment, third, in the test group estimate the heterogenic effects to check if the true outcomes are equal to the estimated ones.

#### Why do we need to use the honest.causalTree function?

Unlike adaptive trees, honest trees divide the data in such a way that one part trains the data and the other tests. The use of honest trees helps not to fall into the problem of overfitting the outcomes. A honest estimation results with tree structure built on training sample (including cross validation) and leaf treatment and ect estimates taken from estimation sample. In this sense, we use the honest.causalTree function to use the split and estimation groups, taking into account the treatment. In this way, with the split data we can build the tree and with estimate data we can calculate the heterogeneous effects.

Let’s remember that there is also another way which is the "adaptive" (which executes the efficient way of how the trees should be divided and with that same splitted data we execute to get the heterogeneous effects of data).
