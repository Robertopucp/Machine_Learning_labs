## Prediction Question

```python
# Import packages for OLS regression
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# Models required
basic = 'lwage ~ sex + exp1 + education + mw + so + we + occ2+ ind2'
Y = 'lwage ~  exp1 + education + mw + so + we + occ2+ ind2'
D = 'sex   ~  exp1 + education + mw + so + we + occ2+ ind2'
basic_results_0=smf.ols(Y , data=data)
basic_results = smf.ols(basic , data=data).fit()
print(basic_results.summary()) # estimated coefficients
print( "Number of regressors in the basic model:",len(basic_results.params), '\n')  # number of regressors in the Basic Model
```

##### Note that the basic model consists of $48$ regressors.

```python
# 2. flexible model
flex = 'lwage ~ (exp1 + exp2 + exp3 + exp4 + education + occ2 + ind2 + mw + so + we)**2'
W = 'lwage ~ (exp1 + exp2 + exp3 + exp4 + education + occ2 + ind2 + mw + so + we)**2'
S = 'sex ~ (exp1 + exp2 + exp3 + exp4 + education + occ2 + ind2 + mw + so + we)**2'
flex_results_0 = smf.ols(W , data=data)
flex_results = smf.ols(flex , data=data).fit()
print(flex_results.summary()) # estimated coefficients
print( "Number of regressors in the basic model:",len(flex_results.params), '\n') # number of regressors in the Flexible Model
```

Note that the flexible model consists of $826$ regressors.

## Try Lasso next and Partialling out

```python
# Import relevant packages for lasso 
from sklearn.linear_model import LassoCV
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
# Get exogenous variables from flexible model
X = basic_results_0.exog
X.shape

# Set endogenous variable
lwage = data["lwage"]
lwage.shape
```

```python
sex=data["sex"]
sex
```

```python
alpha = 0.5
```

```python
# Set penalty value = 0.5
#Lasso Regression
lasso = linear_model.Lasso(alpha = alpha)

#Lasso fitted and predict on Lwage ~ Y
lwage_lasso_fitted = lasso.fit(X, data[['lwage']]).predict(X)

#Get residual from previous model
res= data[['lwage']] - lwage_lasso_fitted.reshape(lwage_lasso_fitted.size ,1)

#Lasso fitted and predict on sex ~ D

basic_results_D = smf.ols(D, data=data)
Sex_lasso = lasso.fit( basic_results_D.exog, data[['sex']] ).predict(basic_results_D.exog)

#Get residual from previous model

res1 = data[['sex']] - Sex_lasso.reshape( Sex_lasso.size, 1)
res2 = pd.DataFrame( np.hstack(( res , res1 )) , columns = ['Res','Res1'] )

#Partialling out

lasso_fit_res = smf.ols( formula = 'Res ~ Res1' , data = res2 ).fit()
coef = lasso_fit_res.summary2().tables[1]['Coef.']['Res1']

# coefficient
print('Lasso Regression: Beta of sex coefficient', coef)
```

# Explanation of Sample Splitting

In statistics, you never know what the population looks like, you only have one sample - representative or not - and you can only model according to that sample you have. Splitting is a tool used in data analysis and machine learning, the objective of which is to generalize very well to the population through a model. It consists of randomly separating the sample into two subsets called: train set and test set. It is common to include in the whole train set 75-80% of the total data and on the other hand, 20-25% of towards the test subsample. The importance of this procedure is that it avoids two situations: Overfitting and Underfitting. Overfitting occurs when our machine learns from particular cases and if these do not conform to it, then it will not be able to recognize the new data. On the other hand, underfitting occurs when our model cannot capture the trend, it may be due to the presence of outliers. The steps to follow is to adjust a set of trains, then this model regresses with this new data (test set). And since it's new data, you're basically seeing how your model performs on it.
